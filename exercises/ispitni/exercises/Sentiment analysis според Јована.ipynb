```python
import pandas as pd
```

```python
data = pd.read_csv('/content/IMDB_5.csv')
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(data['review_es'],  data['sentiment'], test_size=0.2)
```

```python
from collections import Counter

print(f"Training class distributions summary: {Counter(Y_train)}")
print(f"Test class distributions summary: {Counter(Y_test)}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)
```

```python
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(Y_test, y_pred))
```

```python
from imblearn.pipeline import make_pipeline as make_pipeline_imb
from imblearn.under_sampling import RandomUnderSampler

model = make_pipeline_imb(TfidfVectorizer(), RandomUnderSampler(), MultinomialNB())
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)
```

```python
print(classification_report_imbalanced(Y_test, y_pred))
```
